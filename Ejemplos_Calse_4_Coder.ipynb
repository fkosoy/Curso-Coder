```python
import pandas as pd
datos = {'Nombre': ['   Carlos', 'Jorge   ', 'Cris'], 'Correo': ['CS@gmail.com', 'JS@hotmail.com', 'CSS@yahoo.com'], 'Pais': ['Mexico', 'Peru', 'Chile']}
dataframe = pd.DataFrame(datos)
print(dataframe)
```

```python
#recorto correo
dataframe['dominio'] = dataframe['Correo'].str.split('@').str[1]
print(dataframe)
```

```python
#reemplazo variable
dataframe['Nombre'] = dataframe['Nombre'].str.replace('Carlos','Hector')
print(dataframe)
```

```python
#elimino espacios
dataframe['Nombre'] = dataframe['Nombre'].str.strip()
print(dataframe)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# creo ejes y valores con arrays de np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(10, 5))
plt.plot(x, y, label= 'Seno de x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Funci√≥n Seno')
plt.legend()
plt.grid(True)
plt.show()
```
